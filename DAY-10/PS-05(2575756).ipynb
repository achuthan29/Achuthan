```python
cast = pd.read_csv('cast.csv',index_col=None)
cast.head()
```

```python
release_dates = pd.read_csv('release_dates.csv',index_col=None)
release_dates.head()
```

# Make a bar plot of the months in which movies with "Christmas" In their title tend to be released in the USA.

```python
christmas_movies = release_dates[release_dates['title'].str.contains('Christmas', case=False, na=False)]
usa_christmas_movies = christmas_movies[christmas_movies['country'] == 'USA']
usa_christmas_movies['release_date'] = pd.to_datetime(usa_christmas_movies['date'])
usa_christmas_movies['release_month'] = usa_christmas_movies['release_date'].dt.month
monthly_counts = usa_christmas_movies['release_month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Christmas Movies Released by Month (USA)')
plt.xlabel('Month')
plt.ylabel('Number of Movies Released')
plt.xticks(rotation=0)
plt.show()
```

# Make a bar plot of the months in which movies whose titles start with "The Hobbit" are released in the USA.

# Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA.

```python
romance_movies = release_dates[release_dates['title'].str.contains('Romance', case=False, na=False)]
usa_romance_movies = romance_movies[romance_movies['country'] == 'USA']
usa_romance_movies['release_date'] = pd.to_datetime(usa_romance_movies['date'])
usa_romance_movies['release_day_of_week'] = usa_romance_movies['release_date'].dt.day_name()
day_of_week_counts = usa_romance_movies['release_day_of_week'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.figure(figsize=(10, 6))
day_of_week_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Day of the Week for "Romance" Movie Releases (USA)')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Movies Released')
plt.xticks(rotation=45)
plt.show()
```

# On which date was each Judi Dench movie from the 1990s released in the USA?


```python
judi_dench_movies = cast[cast['name'] == 'Judi Dench']
usa_releases = release_dates[release_dates['country'] == 'USA']
merged_data = pd.merge(judi_dench_movies, usa_releases, on='title', how='inner')
merged_data['year'] = pd.to_datetime(merged_data['date']).dt.year
movies_1990s = merged_data[(merged_data['year'] >= 1990) & (merged_data['year'] <= 1999)]
movie_dates_1990s = movies_1990s[['title', 'date']]
print(movie_dates_1990s)
```

# In which months do films with Judi Dench tend to be released in the USA?


```python
usa_releases = release_dates[release_dates['country'] == 'USA']
merged_data = pd.merge(judi_dench_movies, usa_releases, on='title', how='inner')
merged_data['release_date'] = pd.to_datetime(merged_data['date'])
merged_data['release_month'] = merged_data['release_date'].dt.month
monthly_counts = merged_data['release_month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Month-wise Distribution of Judi Dench Movie Releases (USA)')
plt.xlabel('Month')
plt.ylabel('Number of Movies Released')
plt.xticks(rotation=0)
plt.show()
```
